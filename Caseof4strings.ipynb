```python
import random

def pick_two_ends(ends):
    end1, end2 = random.sample(ends, 2)
    ends.remove(end1)
    ends.remove(end2)
    return end1, end2

def replace_ends(ends, end1, end2, turn, loop_count):
    if end1[0] == end2[0]:  # Check if ends have the same number
        print("Loop formed with ends:", end1, "and", end2)
        loop_count[turn] += 1
    else:
        corresponding_end1 = (end1[0], 'B' if end1[1] == 'A' else 'A')
        corresponding_end2 = (end2[0], 'B' if end2[1] == 'A' else 'A')
        if corresponding_end1 in ends:
            ends.remove(corresponding_end1)
        if corresponding_end2 in ends:
            ends.remove(corresponding_end2)
        if turn == 1:
            ends.append((4, 'A'))  # New end
            ends.append((4, 'B'))  # New end
            print("No loop formed. New ends added (turn 1):", (4, 'A'), "and", (4, 'B'))
        elif turn == 2:
            ends.append((5, 'A'))  # New end
            ends.append((5, 'B'))  # New end
            print("No loop formed. New ends added (turn 2):", (5, 'A'), "and", (5, 'B'))
        elif turn == 3:
            ends.append((6, 'A'))  # New end
            ends.append((6, 'B'))  # New end
            print("No loop formed. New ends added (turn 3):", (6, 'A'), "and", (6, 'B'))

def simulate():
    # First turn
    ends = [(n, 'A') for n in range(4)] + [(n, 'B') for n in range(4)]
    print("Initial list of ends:", ends)
    loop_count = {1: 0, 2: 0, 3: 0, 4: 0}
    
    end1, end2 = pick_two_ends(ends)
    print("Picked ends (First turn):", end1, "and", end2)
    
    replace_ends(ends, end1, end2, 1, loop_count)
    
    print("Final list of ends (First turn):", ends)

    # Second turn
    end1, end2 = pick_two_ends(ends)
    print("\nPicked ends (Second turn):", end1, "and", end2)
    if end1 in ends:
        ends.remove(end1)
    if end2 in ends:
        ends.remove(end2)
    replace_ends(ends, end1, end2, 2, loop_count)
    
    print("Final list of ends (Second turn):", ends)

    # Third turn
    end1, end2 = pick_two_ends(ends)
    print("\nPicked ends (Third turn):", end1, "and", end2)
    if end1 in ends:
        ends.remove(end1)
    if end2 in ends:
        ends.remove(end2)
    replace_ends(ends, end1, end2, 3, loop_count)
    
    print("Final list of ends (Third turn):", ends)

    # Fourth turn
    print("\nFourth turn:")
    end1, end2 = ends[0], ends[1]
    print("Picked ends (Fourth turn):", end1, "and", end2)
    if end1[0] == end2[0]:  # Check if ends have the same number
        print("Loop formed with ends:", end1, "and", end2)
        loop_count[4] += 1
    else:
        print("No loop formed.")
    ends.remove(end1)
    ends.remove(end2)
    print("Final list of ends (Fourth turn):", ends)

    print("\nTotal loops formed in each turn:", loop_count)
    print("Total number of loops produced:", sum(loop_count.values()))

simulate()
```
